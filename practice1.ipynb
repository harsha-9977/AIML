```python
import numpy as np
```

```python
x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

x,y = np.meshgrid(x,y)

z = 2*x + 5*y -3

print("Grid : ", x.shape, y.shape, z.shape)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')

# z=0
ax1.plot_wireframe(x,y,np.zeros_like(x), linewidth=0.3)

ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")
ax1.set_title("Meshgrid (flat z = 0)")

ax1.view_init(elev=10, azim=70)

plt.show()

```

```python
# import numpy as np
# import matplotlib.pyplot as plt
# import mpl_toolkit.mplot3d import Axes3D
x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)
x,y = np.meshgrid(x,y)

z = 2*x + 5*y - 3

fig2 = plt.figure()
ax2 = fig2.add_subplot(111,projection='3d')

ax2.plot_surface(x,y,z)

ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")
ax2.set_title("z = 2x + 5y - 3")

ax2.view_init(elev=20,azim=40)

plt.show()
```
