```python
# STEP 1: Import and get antonyms using WordNet
import nltk
from nltk.corpus import wordnet as wn
import gensim.downloader as api

nltk.download('wordnet')

def get_antonyms(word):
    antonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            for ant in lemma.antonyms():
                antonyms.add(ant.name())
    return list(antonyms)

# Get antonyms of 'sad'
antonyms_of_sad = get_antonyms("sad")
print("Antonyms of 'sad':", antonyms_of_sad)

# STEP 2: Load GloVe model
model = api.load('glove-wiki-gigaword-100')

# STEP 3: Find similar words to each antonym
print("\nWords similar to antonyms of 'sad':")
for ant in antonyms_of_sad:
    if ant in model:
        print(f"\nTop words similar to '{ant}':")
        for word, score in model.most_similar(ant, topn=5):
            print(f"{word}: {score:.4f}")
    else:
        print(f"'{ant}' not in vocabulary")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors a, b, c
a = np.array([5, 2, 3])
b = np.array([-2, -1, 5])
c = np.array([1, 3, -2])

# 1. Vector perpendicular to both c and b â†’ cross product
perp_cb = np.cross(c, b)

# 2. Vector perpendicular to (a + b + c)
sum_abc = a + b + c
# Pick arbitrary vector and use cross product to find perpendicular
temp_vec = np.array([1, 0, 0]) if not np.allclose(sum_abc / np.linalg.norm(sum_abc), [1, 0, 0]) else np.array([0, 1, 0])
perp_sum_abc = np.cross(sum_abc, temp_vec)

# 3. Vector parallel to b
parallel_to_b = b / np.linalg.norm(b)  # Unit vector in direction of b

# 4. Vector parallel to b and perpendicular to c
# Use projection of b onto a vector orthogonal to c
# Use Gram-Schmidt process
dot_bc = np.dot(b, c)
proj_b_on_c = (dot_bc / np.dot(c, c)) * c
b_perp_to_c = b - proj_b_on_c  # This is in direction of b but orthogonal to c

# Prepare for plotting
vectors = {
    "a": a,
    "b": b,
    "c": c,
    "perp_cb": perp_cb,
    "perp_sum_abc": perp_sum_abc,
    "parallel_to_b": parallel_to_b * 3,  # scale for visibility
    "b_parallel_perp_c": b_perp_to_c
}

# Plot vectors in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

origin = np.zeros(3)
colors = ['r', 'g', 'b', 'm', 'c', 'y', 'orange']

for (name, vec), color in zip(vectors.items(), colors):
    ax.quiver(*origin, *vec, color=color, label=name, arrow_length_ratio=0.1)

ax.set_xlim([-5, 6])
ax.set_ylim([-5, 6])
ax.set_zlim([-5, 6])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Vector Visualization")
ax.view_init(elev=10,azim=40)
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

def plot_3d_surface(z_func, x_range=(-10, 10), y_range=(-10, 10), resolution=200,
                    title='3D Surface Plot', elev=30, azim=45):
    """
    Plots a 3D surface using matplotlib.

    Parameters:
    - z_func: function of two variables (x, y), e.g. lambda x, y: 2*x + 5*y - 3
    - x_range: (min, max) tuple for x-axis
    - y_range: (min, max) tuple for y-axis
    - resolution: number of points per axis
    - title: title of the plot
    - elev: elevation angle (vertical)
    - azim: azimuth angle (horizontal rotation)
    """
    x = np.linspace(*x_range, resolution)
    y = np.linspace(*y_range, resolution)
    X, Y = np.meshgrid(x, y)
    Z = z_func(X, Y)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

    # Set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title(title)

    # Set view angle
    ax.view_init(elev=elev, azim=azim)

    # Add color bar
    fig.colorbar(surf, shrink=0.5, aspect=5)

    plt.tight_layout()
    plt.show()
```

```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load model once globally (to avoid reloading every time)
glove_model = api.load("glove-wiki-gigaword-100")

def plot_words_pca_2d(word_list):
    """Plot given words in 2D using PCA on GloVe embeddings."""
    vectors = []
    valid_words = []

    # Get vectors for words in vocabulary
    for word in word_list:
        if word in glove_model:
            vectors.append(glove_model[word])
            valid_words.append(word)
        else:
            print(f"âš ï¸ Word not found in vocabulary: '{word}'")

    if len(valid_words) < 2:
        print("âŒ Need at least 2 valid words to plot.")
        return

    # Reduce to 2D using PCA
    pca = PCA(n_components=2)
    reduced_vectors = pca.fit_transform(vectors)

    # Plotting
    plt.figure(figsize=(8, 6))
    for i, word in enumerate(valid_words):
        x, y = reduced_vectors[i]
        plt.scatter(x, y, marker='o', color='blue')
        plt.text(x + 0.01, y + 0.01, word, fontsize=12)

    plt.title("2D PCA Projection of Word Embeddings")
    plt.xlabel("PCA 1")
    plt.ylabel("PCA 2")
    plt.grid(True)
    plt.show()
```

```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load GloVe model (only once)
glove_model = api.load("glove-wiki-gigaword-100")

def plot_words_pca_3d(word_list):
    """Plot given words in 3D using PCA on GloVe embeddings."""
    vectors = []
    valid_words = []

    # Get embeddings
    for word in word_list:
        if word in glove_model:
            vectors.append(glove_model[word])
            valid_words.append(word)
        else:
            print(f"âš ï¸ Word not found in vocabulary: '{word}'")

    if len(valid_words) < 3:
        print("âŒ Need at least 3 valid words to plot in 3D.")
        return

    # PCA to 3D
    pca = PCA(n_components=3)
    reduced_vectors = pca.fit_transform(vectors)

    # Plotting
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    for i, word in enumerate(valid_words):
        x, y, z = reduced_vectors[i]
        ax.scatter(x, y, z, marker='o', color='blue')
        ax.text(x + 0.01, y + 0.01, z + 0.01, word, fontsize=10)

    ax.set_title("3D PCA Projection of Word Embeddings")
    ax.set_xlabel("PCA 1")
    ax.set_ylabel("PCA 2")
    ax.set_zlabel("PCA 3")
    ax.view_init(elev=20, azim=60)
    plt.tight_layout()
    plt.show()
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Symbol and function
x = sp.Symbol('x')
f = 3*x**2 - 4*x + 8

# First and second derivatives
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Solve for critical points (where f' = 0)
critical_points = sp.solve(f_prime, x)

# Analyze critical points
for cp in critical_points:
    second_derivative = f_double_prime.subs(x, cp)
    nature = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection"
    y_val = f.subs(x, cp)
    print(f"{nature} at x = {cp}, f(x) = {y_val}")

# --- Plotting ---
# Convert symbolic function to numerical
f_lambdified = sp.lambdify(x, f, 'numpy')

# Range for plotting
x_vals = np.linspace(-2, 3, 400)
y_vals = f_lambdified(x_vals)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x) = 3xÂ² - 4x + 8', color='blue')

# Mark critical point
for cp in critical_points:
    cp_val = float(cp)
    y_val = float(f.subs(x, cp))
    plt.scatter(cp_val, y_val, color='red', zorder=5)
    plt.text(cp_val, y_val + 0.5, f"Min at x={cp_val:.2f}", ha='center', color='red')

plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Function and its Minimum Point")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Data
x = np.array([36, 38, 32, 31, 29])
y = np.array([3.9, 4.2, 3.4, 3.3, 3.0])
n = len(x)

# Initialize parameters
m = 0.0  # slope
c = 0.0  # intercept
lr = 0.0005  # learning rate

# Function to compute MSE
def compute_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Gradient Descent
prev_loss = float('inf')
tolerance = 1e-6
iteration = 0
max_iters = 100000  # safety stop

while iteration < max_iters:
    # Predict
    y_pred = m * x + c
    error = y_pred - y
    loss = compute_loss(y, y_pred)

    # Gradients
    dm = (2/n) * np.sum(x * error)
    dc = (2/n) * np.sum(error)

    # Update weights
    m -= lr * dm
    c -= lr * dc

    # Print step
    print(f"Iter {iteration:4d} | Loss: {loss:.6f} | m: {m:.4f} | c: {c:.4f}")

    # Stop if change in loss is tiny
    if abs(prev_loss - loss) < tolerance:
        break

    prev_loss = loss
    iteration += 1

print("\nâœ… Converged!")
print(f"Final model: y = {m:.4f}x + {c:.4f}")
print(f"Final MSE: {loss:.6f}")
```
