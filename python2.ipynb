```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 120

def plot_fn(fn, title, x=np.linspace(-10,10,400)):
  plt.figure()
  plt.plot(x, fn(x))
  plt.axhline(0, lw=1, color='gray')
  plt.axvline(0, lw=1, color='gray')
  plt.title(title)
  plt.grid(True)
  plt.show()
```

Linear and piecewise linear shapes

```python
#1 Rising Line y = 2x + 1
plot_fn(lambda t : 2*t + 1, "Rising Line y = 2x + 1")
```

```python
#2 Falling line y = -3x + 4
plot_fn(lambda t : -3*t + 4, "Falling line y = -3x + 4")
```

```python
#3 Horizontal y = 5
plot_fn(lambda t: np.full_like(t,5), "Horizontal y = 5")
```

```python
#4 Vertical (special - plotted via axvline)
plt.figure; plt.axvline(2); plt.title("Vertical line x = 2")
plt.axhline(0,lw=1,color='gray'); plt.ylim(-10,10); plt.grid(True); plt.show()
```

```python
#5 ReLUs
plot_fn(lambda t : np.maximum(0,t), "ReLU y = max(0,x)")
```

```python
#6 Shifted ReLU
plot_fn(lambda t : np.maximum(-2,t), "Shifted ReLU y = mac(-2,x)")
```

```python
#7 absolute values
plot_fn(lambda t : np.abs(t), "V-shape y = |x|")
```

```python
#8 Slanted V
plot_fn(lambda t : np.abs(2*t - 3) + 1, "Slanted V y = |2x-3| + 1")
```

```python
#9 piece-wise quadratic -> Linear
plot_fn(lambda t : np.where(t < 4, t**2 + 1, 2*t - 3), "Piecw-wise hybrid x < 4 : x^2+1 else: 2x+3")
```

```python
#6 S-curve cubic
plot_fn(lambda t: t**3 - 7*t**2 + 12*t, "S-curve cubic  y = x - 7x + 12x")
```

 1. Multivariable Surfaces (z = f(x, y))

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')

# Try different Z formulas:
Z = np.sin(np.sqrt(X**2 + Y**2))  # ripple surface

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("z = sin((x + y))")
plt.show()
```

2. Decision Boundaries in 2D

```python
x = y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2 - 4  # Circle boundary (Z=0 is the boundary)

plt.contour(X, Y, Z, levels=[0], colors='red')
plt.title("Decision Boundary: x + y = 4")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

3. Polar Plots & Parametric Curves

```python
theta = np.linspace(0, 2*np.pi, 500)
r = np.sin(2*theta)

plt.polar(theta, r)
plt.title("Polar Plot: r = sin(2)")
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 500)
x = np.sin(3*t)
y = np.sin(4*t)

plt.plot(x, y)
plt.title("Lissajous Curve")
plt.axis('equal')
plt.grid(True)
plt.show()
```
