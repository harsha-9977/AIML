```python
# find out the sum of square of first 20 natural numbers

import numpy as np
import time

start_list = time.time()

square_list = [i**2 for i in range(1,21)]

sum_list = sum(square_list)

end_list = time.time()
time_list = end_list - start_list

start_np = time.time()

arr = np.arange(1,21)
square_arr = arr ** 2
sum_arr = sum(square_arr)

end_np = time.time()
time_np = end_np - start_np

print("LIST : ")
print("List squares : ", square_list)
print("sum of list : ", sum_list)
print("Time : ", time_list)

print("Numpy Array : ")
print("Array squares : ", square_arr.tolist())
print("sum of numpy : ", sum_arr)
print("Time : ", time_np)
```

```python
# number says that sum of first 3 consecutive numver is next number first 3 numbers are 1,2,5

def hida(n):
  li = [1,2,5]

  for i in range(3,n):
    li.append(li[i-3]+li[i-2]+li[i-1])

  return li
print(hida(98))
```

```python
# max pooling
# import numpy as np

matrix = np.random.randint(0, 256, size=(40,40), dtype=np.uint8)
print(f"40 x 40 matrix \n {matrix}")

def pooled_matrix(input_matrix,kernal_size=2,stride=2):
  output_shape = (
      input_matrix.shape[0] // 2,
      input_matrix.shape[1] // 2
  )

  pooled = np.zeros(output_shape, dtype=np.uint8)

  for i in range(0, input_matrix.shape[0], stride):
    for j in range(0, input_matrix.shape[1], stride):
      patch = input_matrix[i:i+kernal_size, j:j+kernal_size]
      if patch.shape == (kernal_size,kernal_size):
        pooled[i//stride,j//stride] = np.max(patch)
  return pooled

pooled_matrixx = pooled_matrix(matrix)
print(f"20 x 20 matrix \n {pooled_matrixx}")
```

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(matrix, cmap='gray')

plt.subplot(1,2,2)
plt.title("Pooled")
plt.imshow(pooled_matrixx, cmap="gray")

plt.tight_layout()
plt.show()
```
